## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# -------------------------------------------
# Make Dataset Analysis Ready (Data Cleaning)
# -------------------------------------------

# Remove duplicate records
df.drop_duplicates(inplace=True)

# Convert datetime columns to proper format
date_columns = ['order_date_time', 'Issue_reported at', 'issue_responded', 'Survey_response_Date']
for col in date_columns:
    df[col] = pd.to_datetime(df[col], errors='coerce')

# Strip whitespace and standardize text columns
text_cols = ['channel_name', 'category', 'Sub-category', 'Customer_City', 
             'Product_category', 'Agent_name', 'Supervisor', 'Manager',
             'Tenure Bucket', 'Agent Shift']

for col in text_cols:
    df[col] = df[col].astype(str).str.strip().str.title()

# Convert numerical columns to numeric if needed
df['Item_price'] = pd.to_numeric(df['Item_price'], errors='coerce')
df['connected_handling_time'] = pd.to_numeric(df['connected_handling_time'], errors='coerce')
df['CSAT Score'] = pd.to_numeric(df['CSAT Score'], errors='coerce')

# Handling missing values
# Option A: Fill with placeholder text or mode for categorical columns
for col in text_cols:
    df[col].fillna("Unknown", inplace=True)

# Option B: Fill with median for numerical values
num_cols = ['Item_price', 'connected_handling_time', 'CSAT Score']
for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)

# Feature Engineering: Response Time + Survey Delay
df['Response_Time_Hours'] = (df['issue_responded'] - df['Issue_reported at']).dt.total_seconds() / 3600
df['Survey_Delay_Days'] = (df['Survey_response_Date'] - df['issue_responded']).dt.days

# Replace negative or unrealistic values if any
df['Response_Time_Hours'] = df['Response_Time_Hours'].apply(lambda x: np.nan if x < 0 else x)
df['Survey_Delay_Days'] = df['Survey_Delay_Days'].apply(lambda x: np.nan if x < 0 else x)

print("✔ Dataset is now cleaned and ready for further analysis!")
````

### What all manipulations have you done and insights you found?

To make the dataset ready for analysis and derive meaningful insights, several data cleaning and transformation steps (manipulations) were performed, followed by exploratory data analysis:

   * Missing categorical fields were filled with `"Unknown"` or the most frequent value.

4. **Standardization & Cleanup of Text Columns**

   * Trimmed whitespaces and ensured consistent text formatting to avoid duplicate categories appearing differently (e.g., “chat” vs “Chat”).

5. **Feature Engineering**

   * Created important operational performance measures:

     * **Response_Time_Hours**
       Measures delay between customer complaint and agent’s first response.
     * **Survey_Delay_Days**
       Indicates how long a customer takes to provide a CSAT response.

6. **Outlier Replacement**

   * Negative or unrealistic time gaps were replaced with `NaN`.
